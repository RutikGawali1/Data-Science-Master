##Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans -
- Scatter Plot:

Uses: Visualize the relationship between two continuous variables. It helps to identify patterns, clusters, or correlations between the variables.

- Line Plot:
Uses: Display the trend or progression of a variable over time or any ordered category. It is useful for showing data trends, comparing multiple lines, or displaying time series data.

- Bar Plot:
Uses: Compare categorical data by displaying the distribution of a quantitative variable across different categories. It is effective for visualizing comparisons and identifying patterns or differences between groups.

- Histogram:
Uses: Show the distribution of a single continuous variable by dividing it into bins and displaying the frequency or count of data points falling into each bin. It helps in understanding the data distribution and identifying outliers.

- Box Plot:
Uses: Visualize the summary statistics of a continuous variable, including the median, quartiles, and any potential outliers. It is useful for comparing distributions and detecting anomalies or extreme values.

##Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn  as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
# Plot the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

##Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
%matplotlib inline
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
# Reshape the data into a pivot table format
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")
```
