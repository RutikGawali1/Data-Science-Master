#Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
Ans -
- Both the t-test and the z-test are statistical hypothesis tests used to compare means of two groups or to assess whether a sample mean significantly differs from a known population mean. The main difference between the two lies in the assumptions made about the data and the underlying population.
- T-test:

 - The t-test is used when the sample size is small (typically n < 30) and the population standard deviation is unknown.
It assumes that the data follow a normal distribution.
The test statistic in the t-test is the t-value, and the degrees of freedom are calculated based on the sample size.
There are two types of t-tests: the independent samples t-test for comparing two independent groups, and the paired samples t-test for comparing two related groups.

 - Example scenario for using a t-test:
Let's say a researcher wants to compare the average scores of two groups of students who received different teaching methods. The researcher has collected two samples of test scores, each from a different group. Since the sample sizes are relatively small and the population standard deviation is unknown, the researcher would use an independent samples t-test to determine if there is a significant difference in the means of the two groups.

- Z-test:

 - The z-test is used when the sample size is large (typically n ≥ 30) or when the population standard deviation is known.
It assumes that the data follow a normal distribution.
The test statistic in the z-test is the z-value, which is obtained by standardizing the sample mean using the population standard deviation or an estimated standard deviation.
The z-test is often used for hypothesis testing involving population proportions.
 - Example scenario for using a z-test:
Suppose a company claims that the average weight of their product packaging is 500 grams, with a known population standard deviation of 20 grams. To test this claim, a quality control manager takes a large random sample of the product packaging and measures their weights. Since the sample size is large, the manager can use a z-test to determine whether the mean weight of the packaging significantly differs from the claimed value of 500 grams.

#Q2: Differentiate between one-tailed and two-tailed tests.
Ans -
- One-tailed and two-tailed tests are types of hypothesis tests used in statistical analysis. They differ in how they define the critical region and the directionality of the hypothesis being tested.

- One-tailed test:
Also known as a directional test, a one-tailed test is used when the hypothesis specifies a specific direction of the effect or difference between groups. The critical region is defined on one side of the distribution curve. The hypothesis is tested to determine whether the sample data significantly supports the idea that the parameter being tested is either greater than or less than a certain value.
- Example:
Hypothesis: A new drug treatment will increase the average test scores of students.

- One-tailed null hypothesis (H0): The new drug treatment has no effect or decreases the average test scores of students.
One-tailed alternative hypothesis (H1): The new drug treatment increases the average test scores of students.
In this example, the critical region would be on one side of the distribution curve, depending on the direction specified in the alternative hypothesis (either to the right or to the left).

- Two-tailed test:
Also known as a non-directional test, a two-tailed test is used when the hypothesis does not specify a particular direction of the effect or difference between groups. The critical region is defined on both sides of the distribution curve. The hypothesis is tested to determine whether the sample data significantly supports the idea that the parameter being tested is different from a certain value (either greater or less).
- Example:
Hypothesis: A new teaching method will have an effect on the average test scores of students.

- Two-tailed null hypothesis (H0): The new teaching method has no effect on the average test scores of students.
Two-tailed alternative hypothesis (H1): The new teaching method has an effect on the average test scores of students (it could be either an increase or a decrease).

#Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
Ans -
- In hypothesis testing, Type 1 and Type 2 errors are two types of mistakes that can occur when making decisions based on statistical significance. These errors are related to accepting or rejecting a null hypothesis (H0) when it is true or false, respectively. They are denoted as follows:

- Type 1 Error (False Positive):
A Type 1 error occurs when we reject a null hypothesis that is actually true. In other words, we conclude that there is a significant effect or difference when, in reality, there is no effect or difference in the population. This error is also known as a "false positive."
Symbolically:

- Null hypothesis (H0): True
Decision: Reject H0
Reality: H0 is true
Probability of Type 1 Error: Denoted as α (alpha), it is the significance level set by the researcher, usually 0.05 (5%) or 0.01 (1%).

- Example scenario for Type 1 Error:
Suppose a pharmaceutical company is testing a new drug to see if it improves the recovery time from a specific illness. The null hypothesis (H0) states that the drug has no effect on recovery time. After conducting the study and analyzing the data, the researchers find a statistically significant difference in recovery time between the treatment group (using the new drug) and the control group (using a placebo). However, in reality, the drug has no effect, and the observed difference is due to random variation or other factors unrelated to the drug. Thus, the researchers commit a Type 1 error by rejecting the null hypothesis when it was true.

- Type 2 Error (False Negative):
A Type 2 error occurs when we fail to reject a null hypothesis that is actually false. In other words, we conclude that there is no significant effect or difference when, in reality, there is a genuine effect or difference in the population. This error is also known as a "false negative."
Symbolically:

- Null hypothesis (H0): False
Decision: Fail to reject H0
Reality: H0 is false
Probability of Type 2 Error: Denoted as β (beta), it depends on the effect size, sample size, and chosen significance level.

- Example scenario for Type 2 Error:
Continuing with the drug example, suppose the pharmaceutical company's study fails to find a statistically significant difference in recovery time between the treatment and control groups. Based on this result, the researchers conclude that the drug has no effect on recovery time. However, in reality, the drug does have a significant effect, but the study was unable to detect it due to small sample sizes, insufficient statistical power, or other limitations. In this case, the researchers commit a Type 2 error by failing to reject the null hypothesis when it was false.

Ans -
The margin of error (MOE) in a confidence interval is a measure of the uncertainty or range of error associated with the sample estimate. It represents the maximum amount by which the sample estimate is likely to differ from the true population parameter, with a certain level of confidence.

In a confidence interval, the margin of error is calculated using the standard error (SE) of the sample estimate and the critical value corresponding to the desired confidence level. The formula for the margin of error is:

The critical value is obtained from the appropriate distribution, such as the Z-distribution for large sample sizes or the t-distribution for small sample sizes, depending on the context.

Now, as for the relationship between sample size and the margin of error:

Larger Sample Size: With a larger sample size, the margin of error tends to decrease. This is because a larger sample provides more information about the population, leading to a more precise estimate of the population parameter. As the sample size increases, the standard error decreases, and therefore, the margin of error becomes smaller.
